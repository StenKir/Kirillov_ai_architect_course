```python
#Задание №1
import numpy as np

matrix_A = np.array([[1, 2, 1], [-2,1,1]])
matrix_B = np.array([[3, 1],[-1,0],[2,4]])
matrix_C = np.array([[-2, 2],[1,-1],[1,2]])

transpose_matrix_A = matrix_A.transpose()
result_1 = transpose_matrix_A + np.dot(matrix_B, 2)
print(result_1)
result_2 = matrix_B + np.dot(matrix_C, 2)
print(result_2)
```

```python
#Задание №3
import numpy as np

vector1 = np.arange(0, 1.5, 0.1)
vector2 = np.arange(1, 7, 1)

matrix1 = vector1.reshape(5, 3)
matrix2 = vector2.reshape(3, 2)
matrix3 = np.dot(matrix1, matrix2)
print(matrix1)
print(matrix2)
print(matrix3)
```
